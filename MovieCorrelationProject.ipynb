```python
# Upload file
from google.colab import files
uploaded = files.upload()

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Optional styling
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,8)

# Read CSV (after upload)
df = pd.read_csv('movies.csv')
df.head()
```

```python
# Lets look at the data
df.head()
```

```python
# Let's see if there is any missing data
# Let's loop through the data and see if there is anything missing
for col in df.columns:
 for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for our colums
df.dtypes
```

```python
# Change data types of columns

# Drop rows with missing values
df.dropna(subset=["budget", "gross"], inplace=True)
df["budget"] = df["budget"].astype(int)
df["gross"] = df["gross"].astype(int)

# Then change the data types
df["budget"] = df["budget"].astype(int)
df["gross"] = df["gross"].astype(int)
df.head()

```

```python
# Create correct year column
df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df= df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df_numerized =df

for col_name in df_numerized.columns:
  if(df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
# For easy visualizations
df_numerized.corr()
```

```python
# Unstacking

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
correlation_matrix = df.corr(numeric_only=True)
sorted_pairs = correlation_matrix.unstack().sort_values(kind="quicksort")
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
